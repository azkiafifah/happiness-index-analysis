## Gabung dataset 2015-2020

```python
datarank = pd.DataFrame(columns = ['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)',
                                   'Family', 'Health (Life Expectancy)', 'Freedom','Trust (Government Corruption)',
                                   'Generosity', 'Year'])
n = [data2015, data2016, data2017, data2018, data2019, data2020]
for i in n:
    datarank = datarank.append(i[['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)',
                                   'Family', 'Health (Life Expectancy)', 'Freedom','Trust (Government Corruption)',
                                   'Generosity', 'Year']], ignore_index = 'True')
```

```python
def color_rank(value):
    if value == 1:
        color = 'yellow'
    elif value == 2:
        color = 'silver'
    elif value == 3:
        color = 'moccasin'
    else:
        color = 'default'
    return 'background-color: %s' % color


def highlight_max(data, color='yellow'):
    '''
    highlight the maximum in a Series or DataFrame
    '''
    attr = 'background-color: {}'.format(color)
    data = data.astype(float)
    if data.ndim == 1: 
        is_max = data == data.max()
        return [attr if v else '' for v in is_max]
    else:  
        is_max = data == data.max().max()
        return pd.DataFrame(np.where(is_max, attr, ''),
                            index=data.index, columns=data.columns)
    
```

1. Apakah kekayaan dapat mempengaruhi happiness index suatu negara? Semakin kaya semakin bahagia?

```python
datarank.info()
```

```python
#Linear Regression Happiness score and GDP Per Capita 

from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

reg = LinearRegression()
X = datarank[['Economy (GDP per Capita)']]
Y = datarank['Happiness Score']

reg.fit(X,Y)
Yp = reg.predict(X)

plt.figure()
plt.scatter(X['Economy (GDP per Capita)'],Y, color='blue', marker='o')
plt.scatter(X['Economy (GDP per Capita)'],Yp, color='red', marker='+', alpha=0.5)
plt.show()

#Nilai MSE dan RMSE

mse = mean_squared_error(Y,Yp)
print('GDP per capita as X, MSE = ',str(mse))
print('GDP per capita as X, RMSE = ',str(np.sqrt(mse)))
```
